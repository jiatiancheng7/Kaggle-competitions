```
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import Lasso
import numpy as np
import pandas as pd
boston_data = datasets.load_boston()
X = boston_data.data
Y = boston_data.target

SSE=[]
X_train,X_test,y_train,y_test = train_test_split(X,Y,test_size=0.3,random_state=31)

## UserWarning: With alpha=0, this algorithm does not converge well. You are advised to use the LinearRegression estimator.

for i in range(0,101,1):
  model = Lasso(alpha=i)  
  model.fit(X_train,y_train)
  SSE.append(model.score(X_test,y_test))
print(SSE)


plt.plot(range(0,101,1), SSE)
plt.show()

```

```
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import Ridge
import numpy as np
import pandas as pd
boston_data = datasets.load_boston()
X = boston_data.data
Y = boston_data.target

SSE=[]
X_train,X_test,y_train,y_test = train_test_split(X,Y,test_size=0.3,random_state=31)
# print(X_train)
# print(y_train)

for i in range(101):
  model = Ridge(alpha=i,normalize=True)  
  model.fit(X_train,y_train)
  SSE.append(model.score(X_test,y_test))
print(SSE)


plt.plot(range(101), SSE)
plt.show()

```

```
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import ElasticNet
import numpy as np
import pandas as pd
boston_data = datasets.load_boston()
X = boston_data.data
Y = boston_data.target
X_train,X_test,y_train,y_test = train_test_split(X,Y,test_size=0.3,random_state=31)
model = ElasticNet(alpha=1, l1_ratio=0.7, normalize=False)
model.fit(X_train,y_train)
print("SSE",model.score(X_test,y_test))
print("coef",model.coef_)
print("intercept",model.intercept_)
```
